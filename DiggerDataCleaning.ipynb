## Attempt to implement shift and delete.



The main goal of this project is to extract the car makes, models and number of users, as mentioned in the beginning.

Looking at the above data, some values have numbers before the car make. 

Goal is to delete the numeric values in the row and shift the rest of the data in that same row to the left.

This will ensure that the 3rd and 4th columns will have the car makes and car models respectively.




Reference : [Pandas. A pretty way to delete cell and shift left others in row?

Reference : [Python Pandas Dataframe, remove all rows where 'None' is the value in any column
](https://stackoverflow.com/questions/45512763/python-pandas-dataframe-remove-all-rows-where-none-is-the-value-in-any-column)

````python
# delete rows that have null values
df_final_hire_sale = df_final_hire_sale.mask(df_final_hire_sale.eq('None')).dropna()
````

Reference : [Repeating values in a "group by" pandas dataframe](https://stackoverflow.com/questions/36676800/repeating-values-in-a-group-by-pandas-dataframe)

````python
# Group data based on values in columns 'Value1' and 'Value2' 
# this groups them based on the car make and whether thay are for sale or for hire
df_final_hire_sale = df_final_hire_sale.groupby(['Value1','Value2'], as_index=False)['Users'].sum()
````
